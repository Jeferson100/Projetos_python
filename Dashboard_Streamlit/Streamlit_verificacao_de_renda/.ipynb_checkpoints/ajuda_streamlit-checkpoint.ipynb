```python
import streamlit as st
import plotly.express as px

# Suponha que vocÃª jÃ¡ tenha a variÃ¡vel "filtered_data" com os dados filtrados

# Calcular a contagem de valores
contagem_valores = data['numero_da_educacao'].value_counts().sort_index().values

# Adicionar valores formatados como rÃ³tulos
labels = [f"{x}\n{n}" for x, n in zip(data['numero_da_educacao'].value_counts().sort_index().index, contagem_valores)]

    
# Criar o grÃ¡fico de histograma
fig = px.histogram(
    data, 
    x='numero_da_educacao',
    labels={'numero_da_educacao': 'Anos de estudo'},
    template="seaborn",
    )
    
 # Adicionar os rÃ³tulos como textos personalizados nas barras
fig.update_traces(text=data.numero_da_educacao.value_counts().sort_index().values, textposition='outside')
    
# Nomear o eixo Y
fig.update_yaxes(title_text='FrequÃªncia')
    
# Mostrar o grÃ¡fico no Streamlit
fig.show(height = 200) 
```

```python
ocupacao = data.ocupacao.value_counts().head(10)
st.subheader("Empregos mais comuns")
fig = px.bar(x = ocupacao.index, 
                 y = ocupacao.values,
                 color_discrete_sequence=["black"],
                 template = "seaborn",
                 labels={'x': 'OcupaÃ§Ã£o', 'y': 'FrequÃªncia'},
                 text= ocupacao.values)
fig.update_yaxes(title_text='FrequÃªncia')
fig.show(use_container_width=True, height = 200)
```

```python
import pandas as pd

# Suponha que vocÃª jÃ¡ tenha a variÃ¡vel "renda_raca" com os dados
# Transformar os dados para ter as rendas como colunas

renda_raca = data.groupby(['Raca', 'Renda']).size().reset_index(name='Contagem')
pivot_renda_raca = renda_raca.pivot(index='Raca', columns='Renda', values='Contagem').fillna(0)
pivot_renda_raca.columns = pivot_renda_raca.columns.str.strip()
# Calcular as porcentagens
pivot_renda_raca['Porcentagem_maior'] = (pivot_renda_raca['Maior que 50 mil.'] / pivot_renda_raca.sum(axis=1)).round(2)
pivot_renda_raca['Porcentagem_Menor'] = (pivot_renda_raca['Menor ou igual a 50 mil.'] / pivot_renda_raca.sum(axis=1)).round(2)
```

```python
pivot_renda_raca.index = ['<b>' + str(i) + '</b>' for i in pivot_renda_raca.index]
```
