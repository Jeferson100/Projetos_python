```python
# Configurar o serviço
servico = webdriver.chrome.service.Service(ChromeDriverManager().install())

# Iniciar o navegador com o serviço
navegador = webdriver.Chrome(service=servico)
navegador.get('https://stooq.com/t/?i=515&v=0')

# Localizar o elemento pelo XPath inserindo nome ao login
navegador.find_element(By.XPATH, '/html/body/table/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[2]/tbody/tr/td[2]/table/tbody/tr/td/table/tbody/tr/td[2]/input[1]').send_keys('jefersonsehnem')
#elemento.send_keys('jefersonsehnem'), inserinod a senha
navegador.find_element(By.XPATH, '/html/body/table/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[2]/tbody/tr/td[2]/table/tbody/tr/td/table/tbody/tr/td[2]/input[2]').send_keys('Ce97153844')
##Clicando no botao para fazer login
navegador.find_element(By.XPATH, '/html/body/table/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[2]/tbody/tr/td[2]/table/tbody/tr/td/table/tbody/tr/td[5]/input').click()
navegador.implicitly_wait(10)
#tabela = navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[2]/tbody')
#html_da_tabela = tabela.get_attribute('outerHTML')
#soup = BeautifulSoup(html_da_tabela, 'html.parser')
lista_tick = []
lista_nome = []
contador = 0
proxima_pagina = '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a1'
while contador < 37:
    tabela = navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[2]/tbody')
    html_da_tabela = tabela.get_attribute('outerHTML')
    soup = BeautifulSoup(html_da_tabela, 'html.parser')
    # Encontre todas as tags <a> com o atributo href começando com "q/?s=" 
    tags_a = soup.find_all('a', href=lambda href: href and href.startswith("q/?s="))
    for a_tag in tags_a:
        #print(a_tag.text)  # Imprime o texto dentro da tag <a>
        lista_tick.append(a_tag.text)

    # Encontre as tags <td> com o atributo id igual a "f10"
    tags_td = soup.find_all('td', id="f10")
    for td_tag in tags_td:
        #print(td_tag.text)  # Imprime o texto dentro da tag <td>
        lista_nome.append(td_tag.text)
        
    if contador == 0:
        navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a[1]').click()
    if contador == 36:
        navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/font').click()
    else:
        navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a[3]').click()
    contador = contador+1
df_nyse_stocks = pd.DataFrame()
df_nyse_stocks['ticker'] = lista_tick
df_nyse_stocks['name'] = lista_nome
df_nyse_stocks.drop_duplicates('ticker').dropna(inplace=True)
df_nyse_stocks.to_csv('df_nyse_stocks.csv')
```

```python
# Configurar o serviço
servico = webdriver.chrome.service.Service(ChromeDriverManager().install())

# Iniciar o navegador com o serviço
navegador = webdriver.Chrome(service=servico)
navegador.get('https://stooq.com/t/?i=516')

# Localizar o elemento pelo XPath inserindo nome ao login
navegador.find_element(By.XPATH, '/html/body/table/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[2]/tbody/tr/td[2]/table/tbody/tr/td/table/tbody/tr/td[2]/input[1]').send_keys('jefersonsehnem')
#elemento.send_keys('jefersonsehnem'), inserinod a senha
navegador.find_element(By.XPATH, '/html/body/table/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[2]/tbody/tr/td[2]/table/tbody/tr/td/table/tbody/tr/td[2]/input[2]').send_keys('Ce97153844')
##Clicando no botao para fazer login
navegador.find_element(By.XPATH, '/html/body/table/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[2]/tbody/tr/td[2]/table/tbody/tr/td/table/tbody/tr/td[5]/input').click()

navegador.implicitly_wait(10)
#tabela = navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[2]/tbody')
#html_da_tabela = tabela.get_attribute('outerHTML')
#soup = BeautifulSoup(html_da_tabela, 'html.parser')
lista_tick = []
lista_nome = []
contador = 0
#proxima_pagina = '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a1'
try:
    while contador < 48:
        tabela = navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[2]/tbody')
        html_da_tabela = tabela.get_attribute('outerHTML')
        soup = BeautifulSoup(html_da_tabela, 'html.parser')
        # Encontre todas as tags <a> com o atributo href começando com "q/?s=" 
        tags_a = soup.find_all('a', href=lambda href: href and href.startswith("q/?s="))
        for a_tag in tags_a:
            #print(a_tag.text)  # Imprime o texto dentro da tag <a>
            lista_tick.append(a_tag.text)

        # Encontre as tags <td> com o atributo id igual a "f10"
        tags_td = soup.find_all('td', id="f10")
        for td_tag in tags_td:
            #print(td_tag.text)  # Imprime o texto dentro da tag <td>
            lista_nome.append(td_tag.text)
        
        if contador == 0:
            navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a[1]').click()
        if contador == 47:
            navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a[3]').click()
        else:
            navegador.find_element(By.XPATH, '/html/body/table[1]/tbody/tr[2]/td[2]/table/tbody/tr/td/table/tbody/tr/td/table[5]/tbody/tr/td/table/tbody/tr[2]/td[1]/table[3]/tbody/tr/td/a[3]').click()
            
        contador=contador + 1

except:   
    df_nasdaq_stock = pd.DataFrame()
    df_nasdaq_stock['ticker'] = lista_tick
    df_nasdaq_stock['name'] = lista_nome
    df_nasdaq_stock.drop_duplicates('ticker').dropna(inplace=True)
```
